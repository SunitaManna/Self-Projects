About the Dataset

The dataset contains the top 250 rated movies on IMDB as of 2021, offering insights into popular and highly rated films. Analyzing this dataset can provide valuable information on movie industry trends and popular genres.

Dataset link: "https://www.kaggle.com/datasets/rajugc/imdb-top-250-movies-dataset"

```python
# Importing necessary Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading the Dataset / CSV File

df=pd.read_csv("IMDB Top 250 Movies.csv")
df.head()
```

```python
# Displaying the in detail information of the Dataset i.e. Rows, Columns, Null Values, Datatypes

df.info()

# Insights: 1). The Dataset contain 13 Columns and 250 Rows
#           2). The Dataset does not contain anyy Null Values.
#           3). Most columns have obbject datatype 2 columns have int64 and 1 column have float64.
```

```python
# Checking the Unique Values in Certificate Column

df["certificate"].unique()
```

```python
# Q1. In which year highest number of movies were released.

df["year"].value_counts()

# Insights :- In the year 1995 highest number of movies were released i.e. 8
```

```python
# Q2. Find the directors and genre of The Dark Knight 

df[df["name"]=="The Dark Knight"][["directors","genre"]]

# Insights :- Director of The Dark Knight is Christopher Nolan and Genre is Action,Crime,Drama
```

```python
# Q3. Find the movies of the directors named Sidney Lumet and writers is Reginald Rose.

df[(df["directors"]=="Sidney Lumet") & (df["writers"]=="Reginald Rose")]
```

```python
# Q4. Display all the movies name and writers of the year 2004

df[df["year"]==2004][["name","writers"]]
```

```python
grp=df.groupby(by="certificate")
grp
```

```python
# Q5. Display all the movies with Approved certificate

grp.get_group("Approved")
```

```python
# Q6. Find out directors whose movie got passed certificate

grp.get_group("Passed")["directors"].value_counts()
```

```python
# Q7. Find rating and box_office of David Lean's Lawrence of Arabia

df[(df["name"]=="Lawrence of Arabia") & (df["directors"]=="David Lean")][["rating","box_office"]]
```

```python
# Q8. Display top 20 directors movies released according to the year

plt.figure(figsize=(8,5))
sns.barplot(data=df,x="year",y=df["directors"].head(20))
plt.title("Movie Released")
plt.xlabel("Year")
plt.ylabel("Directors")
plt.xlim(1900,2050)
plt.grid()
plt.show()

# Insights :- Maximum no of movies released are directed by Christopher Nolan.
```

```python
# Q9. Ratio of top 20 movie released on basis of their certification

plt.figure(figsize=(7,5))
colors=sns.color_palette('pastel')
df["certificate"].head(20).value_counts().plot.pie(autopct="%1.0f%%",startangle=45,colors=colors,explode=[0.1,0,0,0,0,0])
plt.title("Certification of Movie Released")
plt.show()

# Insights :- Maximum movies released are of 'R' certification. 
```

```python
# Q10. Find the rating of movies according to their certificates

plt.figure(figsize=(9,5))
sns.barplot(data=df,x="certificate",y="rating")
plt.title("Rating by Certificate")
plt.ylim(6,10)
plt.xticks(rotation=90)
plt.grid()
plt.show()

# Insights :- Highest rating is given to TV-MA, 2nd Highest rating is given to 18+ movie certificates. 
```

```python
# Q11. Find the ratings of a movie where most of the people have rated. 

plt.figure(figsize=(7,5))
plt.hist(x=df["rating"],bins=10)
plt.title("Movie Ratings")
plt.xlabel("Rating")
plt.show()

# Insights :- Most people have rated between the range of 8.0 to 8.6
```

```python
# Q12. Find Rating, Budget and Box_office of The Lord of the Rings: The Return of the King 
#directed by Peter Jackson released in 2003

df[(df["name"]=="The Lord of the Rings: The Return of the King") & 
(df["directors"]=="Peter Jackson") & (df["year"]==2003)][["rating","budget","box_office"]]
```

```python
# Q13. Find Year, Rating and Box_office of Passed certificate.

grp.get_group("Passed")[["year","rating","box_office"]]
```

```python
# Q14. Find name, directors, writers and box_office of Crime,Drama released in 1995

df[(df["genre"]=="Crime,Drama") & (df["year"]==1995)][["name","directors","writers","box_office"]]
```

```python
# Q15. Find name, rating, certificate and box_office movies released in 2009

df[df["year"]==2009][["name","rating","certificate","box_office"]]
```
